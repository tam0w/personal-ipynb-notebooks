```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.DataFrame({'data':[1,2,3,4],'data2':[0,4,0,1]})

fig, ax = plt.subplots(3,1)
ax[0].plot(df['data'], df['data2']) 
```

```python
labels = ['A', 'B', 'C', 'D']
x = np.arange(len(labels))
width = 0.4
plt.bar(x-width/2, [20,25,40,10], width=0.4)
plt.bar(x+width/2, [20,25,30,20], width=0.4)
plt.xticks(x)
ax = plt.gca()
ax.set_xticklabels(labels)
```

```python
x = np.arange(3)
bars1 = [10,20,30]
bars2 = [15,25,30]
bars3 = [20,10,15]
bars4 = [0,20,40]
plt.bar(x,bars1)
plt.bar(x,bars2, bottom=bars1)
plt.bar(x,bars3, bottom=np.add(bars1,bars2))
```

```python
x = np.arange(3)
bars1 = [40,20,30]
bars2 = [15,25,30]
bars3 = [20,10,15]
bars4 = [0,20,50]
plt.bar(x,bars1)
plt.bar(x,bars2, bottom=bars1)
plt.bar(x,bars3, bottom=np.add(bars1,bars2))
plt.bar(x,bars4, bottom=np.add(bars2, bars3))
```
